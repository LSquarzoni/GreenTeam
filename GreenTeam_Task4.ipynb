```python
# Importing packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pyarrow
```

We are still working with the month of January 2021 for the power dataset

```python
path_P = "../21-01/year_month=21-01/plugin=ipmi_pub/metric=total_power/a_0.parquet"
dataset_P = pd.read_parquet(path_P, engine='pyarrow')

# Casting of node values into integers
dataset_P['node'] = dataset_P['node'].astype(int)

dataset_P
```

```python
# As usual I need to order by datetime the samples I have
dataset_P['timestamp'] = pd.to_datetime(dataset_P['timestamp'])

dataset_P = dataset_P.sort_values(by='timestamp')

dataset_P
```

```python
# Pivot the DataFrame
pivot_df = dataset_P.pivot(index='timestamp', columns='node')

# If you want to reset index (optional)
pivot_df.reset_index(inplace=True)

# Interpolate missing values using linear interpolation for each column
#for column in pivot_df.columns:
    #pivot_df[column].interpolate(method='linear', inplace=True)

# Display the resulting DataFrame with missing values filled using linear interpolation
pivot_df
```

```python
# Assuming 'datetime' column is already in datetime format
pivot_df['timestamp'] = pd.to_datetime(pivot_df['timestamp'])

# Set 'datetime' column as the index
pivot_df.set_index('timestamp', inplace=True)

# Resample the DataFrame on an hourly basis and calculate the sum
hourly_sum = pivot_df.resample('1H').sum()

# Display the resulting DataFrame
hourly_sum
```

```python
# Iterate over each column (node) in the DataFrame
for node in hourly_sum.columns:
    # Replace the values in the original DataFrame with the hourly sum values
    pivot_df[node] = hourly_sum[node]

# Convert the index (assuming it's the timestamp) to datetime
pivot_df.index = pd.to_datetime(pivot_df.index)

# Drop rows where minutes and seconds are not zero
pivot_df = pivot_df.drop(pivot_df.index[(pivot_df.index.minute != 0) | (pivot_df.index.second != 0)])

# Display the resulting DataFrame with only rows corresponding to exact hours
pivot_df
```

```python
# Interpolate NaN values in each column using linear interpolation
pivot_df.interpolate(method='linear', inplace=True)

# Display the resulting DataFrame with NaN values filled using linear interpolation
pivot_df
```
