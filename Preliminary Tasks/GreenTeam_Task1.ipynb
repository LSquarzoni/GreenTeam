```python
# Importing packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
path = "../IT-NO_2021_hourly.csv"
```

```python
# Dataset import
dataset = pd.read_csv(path)
display(dataset)
```

```python
dataset.columns
```

```python
# We remove the unnecessary data columns
dataset = dataset.drop(columns=['Zone Name', 'Zone Id', 'Data Source', 'Data Estimated', 'Data Estimation Method'])
dataset.head()
```

```python
# Distribution of data in our dataset
dataset.hist(bins=50, xlabelsize=10)
```

From these simple histograms we can gather some initial information related to our data, for example it is clear how the percentage of renewable energy mainly floats around 30 to 40%, while the low carbon energy percentage is more close to 50 to 60%.

```python
# We want to study the correlation of the values present in the dataset
datacorr = dataset.iloc[:, 2:].corr()
datacorr
```

```python
# Mean values of each KPI
print(dataset.describe(include='all').loc['mean'].loc['Carbon Intensity gCO₂eq/kWh (direct)'])
print(dataset.describe(include='all').loc['mean'].loc['Carbon Intensity gCO₂eq/kWh (LCA)'])
print(dataset.describe(include='all').loc['mean'].loc['Low Carbon Percentage'])
print(dataset.describe(include='all').loc['mean'].loc['Renewable Percentage'])
```

```python
# We want to see important data related to our dataset
dataset.describe()
```

```python
# Here we substitute the NaN values present with the mean value of the respective KPI
dataset['Carbon Intensity gCO₂eq/kWh (direct)'] = dataset['Carbon Intensity gCO₂eq/kWh (direct)'].fillna(dataset.describe(include='all').loc['mean'].loc['Carbon Intensity gCO₂eq/kWh (direct)'])
dataset['Carbon Intensity gCO₂eq/kWh (LCA)'] = dataset['Carbon Intensity gCO₂eq/kWh (LCA)'].fillna(dataset.describe(include='all').loc['mean'].loc['Carbon Intensity gCO₂eq/kWh (LCA)'])
dataset['Low Carbon Percentage'] = dataset['Low Carbon Percentage'].fillna(dataset.describe(include='all').loc['mean'].loc['Low Carbon Percentage'])
dataset['Renewable Percentage'] = dataset['Renewable Percentage'].fillna(dataset.describe(include='all').loc['mean'].loc['Renewable Percentage'])
```

```python
display(dataset)
```

```python
# Box plot to see if there are outliers
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
g = sns.boxplot(data = dataset,linewidth=2.5,ax=ax)
g.set_yscale("log")
plt.title("Box plot of the outliers")
plt.show()
```

Since there is a huge amount of data in the dataset, it is used a data visualization techniques to inspect the data’s distribution and verify the presence of outliers.

- Carbon Intensity gCO₂eq/kWh (direct): no outliers are observed in the first feature of the dataset. This means that the values of this feature fall within a reasonable range without any extreme values.

- Carbon Intensity gCO₂eq/kWh (LCA): outliers are present in this feature, specifically located beyond the third quartile (Q3). These outliers indicate values that are significantly higher than the majority of the data points, suggesting some extreme values in carbon intensity.

- Renewable Percentage: similar to carbon intensity, outliers are found in this feature beyond the third quartile (Q3). This implies that there are data points with renewable percentages much higher than the majority of the dataset.

- Low Carbon Percentage: outliers in this feature are located below the minimum value, falling below the first quartile (Q1). This suggests that there are data points with low carbon percentages significantly lower than the majority of the dataset.

```python
correlations_absolute = abs(datacorr)
kwargs = {"shrink": 0.82}

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations_absolute, vmin=0, vmax = 1)
fig.colorbar(cax, **kwargs)
ticks = np.arange(0, 8, 1)
plt.show()
```

```python
#print("The correlation matrix is:", datacorr)
#relation_index = datacorr.index
#print(relation_index)

sns.heatmap(datacorr,annot=True)
plt.title("Plot of the correlation matrix")
plt.show()
```

This correlation matrix is used to compare the coefficients of correlation between different features of the dataset. In fact it allows to visualize how much (or how little) correlation exists between different variables just by examing the color intensity of the plot. The legend of the plot allows to analyze the strength of the correlation: the values are ina a range from -1 to 1: 
- A value close to 1 indicates a strong positive correlation, while a value close to -1 indicates a strong negative correlation.
- A value close to 0 suggests a lack of correlation or a very weak correlation.

In our case:
- *Carbon Intensity gCO₂eq/kWh (direct)*: highly correlated with all other features, indicating a strong relationship with other variables.
- *Carbon Intensity gCO₂eq/kWh (LCA)*: strong positive correlation with the first feature, and a strong negative correlation with the other two features, suggesting different relationships with different variables.
- *Low Carbon Percentage*: strong negative correlation with the two Carbon Intensity features, and a strong correlation with the Renewable Percentage. This implies that as one variable increases, the other tends to decrease, and vice versa.
- *Renewable Percentage*: strong negative correlation with the two Carbon Intensity features and a strong correlation with the Low Carbon Percentage. This suggests an inverse relationship with the Carbon Intensity features and a positive relationship with the Low Carbon Percentage.

```python
#Datetime on x-axis and Carbon Intensity on y-axis

dataset['Datetime (UTC)'] = pd.to_datetime(dataset['Datetime (UTC)'])
plt.figure(figsize=(10, 6))
plt.plot(dataset['Datetime (UTC)'], dataset['Carbon Intensity gCO₂eq/kWh (direct)'], label='Carbon Intensity gCO₂eq/kWh (direct)')
plt.plot(dataset['Datetime (UTC)'], dataset['Carbon Intensity gCO₂eq/kWh (LCA)'], label='Carbon Intensity gCO₂eq/kWh (LCA)')

plt.xlabel('Datetime')
plt.ylabel('Values')
plt.title('Plot of Carbon Intensity over time')
plt.legend()
plt.grid(True)
plt.show()
```

Finally, looking at the plot of Carbon Intensity emission over the span of a year, we can highlight in which periods and seasons the emissions increase or decrease: for example, it's evident that during the colder months the production of polluting gasses rises in a significant way with respect to the rest of the year.

```python
# We want also to zoom in to look better at the trend within a month
plt.figure(figsize=(10, 6))
plt.plot(dataset['Datetime (UTC)'], dataset['Carbon Intensity gCO₂eq/kWh (direct)'], label='Carbon Intensity gCO₂eq/kWh (direct)')
plt.plot(dataset['Datetime (UTC)'], dataset['Carbon Intensity gCO₂eq/kWh (LCA)'], label='Carbon Intensity gCO₂eq/kWh (LCA)')

plt.xlabel('Datetime')
plt.ylabel('Values')
plt.title('Plot of Carbon Intensity over time')
plt.legend()
plt.grid(True)
plt.xlim(pd.Timestamp('2021-06'), pd.Timestamp('2021-07'))

plt.show()
```

Going deeper we can clearly see a pattern also at week level, meaning that the emissions change with respect to the day we are considering.
